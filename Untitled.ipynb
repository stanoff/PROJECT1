**DATA CLEANING AND PREPARATION**

This involves checking data to check  if there exists any possible problems that makes data unsuitable for analysis.The problems may include missing data , duplicated data , outliers and inconsistencies in the data.Data cleaning and preparation done is shown below;


```python
#Cleaning 'bom.movie_gross.csv'
#df_movie_gross
#For quick preview of features with missing data
print(df_movie_gross.info()) 
#check for duplicate entries
print(df_movie_gross.duplicated().value_counts())
#remove duplicates
print(df_movie_gross.drop_duplicates())
# Drop irrelevant columns
#print(df_movie_gross = df_movie_gross.drop(['studio'], axis=1))
# Drop rows with missing values in relevant columns
df_movie_gross = df_movie_gross.dropna(subset=['domestic_gross', 'foreign_gross', 'year'])
#convert domestic_gross column as an integer
df_movie_gross['domestic_gross'] = df_movie_gross['domestic_gross'].astype(int).astype(str)
#convert foreign_gross column as integer
#df_movie_gross['foreign_gross'] = df_movie_gross['foreign_gross'].astype(int).astype(str)
#create another column called total_gross where domestic_gross and foreign_gross are summed
df_movie_gross['total_gross'] = (df_movie_gross['domestic_gross'] + df_movie_gross['foreign_gross'])
# Convert the "total_gross" column to numeric
df_movie_gross['total_gross'] = pd.to_numeric(df_movie_gross['total_gross'], errors='coerce')

# Divide the numeric values in the "total_gross" column by 1,000,000
df_movie_gross['total_gross'] = df_movie_gross['total_gross'] / 1000000

# Inspect the cleaned dataset
print("df_movie_gross:")
print(df_movie_gross.head())
print(df_movie_gross.info())
print(df_movie_gross.describe())
```

```python
#Cleaning 'title.basics.csv'
#Drop unnecessary columns using drop()
df_title_basics.drop(['original_title','start_year','runtime_minutes'], axis = 1,inplace = True)
#Remove rows with missing genres
df_title_basics = df_title_basics[df_title_basics['genres'].notna()]
print(df_title_basics)
print(df_title_basics.head())
print(df_title_basics.info())
print(df_title_basics.describe())
```

```python
#Cleaning 'title.ratings.csv'
#Drop unnecessary columns
df_ratings.drop(['numvotes'],axis = 1,inplace = True)
print(df_ratings)
print(df_ratings.head())
print(df_ratings.info())
print(df_ratings.describe())
```

```python
#join 'df_title_basics' and 'df_ratings' to get one dataframe
df_imdb = df_title_basics.join(df_ratings, on='tconst')
print(df_imdb)
```

```python
df_imdb.rename(columns = {'primary_title':'title'},inplace = True)
df_imdb.dropna(inplace = True)
print(df_imdb)
```

```python
#create another dataframe with the columns averagerating,total_gross,year
# Create a smaller dataframe of movies with at least a rating of 7.0 and total_gross of at least 27.9mil and year not less than 2012
#df_created = [(df_ratings['averagerating'] >= 7) & (df_movie_gross['total_gross'] >= 27.9) & (df_movie_gross['year'] > 2012)]
#print(df_created)
```

```python
# split the genre column for analysis, so that each genre is independent
moviedf_small = df_imdb.assign(genres = df_imdb['genres'].str.split(',')).explode('genres')
print(moviedf_small)
```

```python
#Create a new dataframe with the count of movies per genre
#genre_counts = df_imdb['genres'].value_counts().reset_index()
#genre_counts.columns = ['genres', 'count']
#print(genre_counts)

genre_counts = (pd.DataFrame(moviedf_small.groupby('genres', as_index = False).title.nunique())).sort_values(
    by = 'title', ascending = False)
print(genre_counts)
```

```python
#create another dataframe with the columns averagerating,total_gross,year
# Assuming the DataFrames are named df_movie_gross and moviedf_small

# Select the desired columns from each DataFrame
df1 = df_movie_gross[['total_gross', 'year']]
df2 = moviedf_small[['averagerating']]

# Combine the columns into a new DataFrame
df_created = pd.concat([df1, df2])

# Display the new DataFrame
print(df_created)

# Filter the DataFrame based on the conditions
filtered_df = df_created[(df_created['averagerating'] > 7.0) & (df_created['total_gross'] > 27.9) & (df_created['year'] > 2012)]

# Display the filtered DataFrame
print(filtered_df)




# Create a smaller dataframe of movies with at least a rating of 7.0 and total_gross of at least 27.9mil and year not less than 2012
#df_created = [(df_ratings['averagerating'] >= 7) & (df_movie_gross['total_gross'] >= 27.9) & (df_movie_gross['year'] > 2012)]
#print(df_created)
```

**EXPLORATORY DATA ANALYSIS**

This involves examining and understanding the structures, patterns and relationships within dataset.It also involves calculating basic descriptive statistics, creating a range of plots like box plot, histogram, scatter plot and other interactive visualizations to focus on specific segments of data.For this case we will perform exploratory data analysis on the joined data set 'df_imdb'

Below is a histogram showing the distribution of movie ratings


```python
#Histogram of movie ratings
plt.figure(figsize = (8,6))
sns.histplot(data = df_imdb, x = 'averagerating',bins = 20,kde =True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

For example, a histogram might show a normal distribution, indicating that most movies receive ratings around a central value, with fewer movies receiving extremely high or low ratings. This suggests that the overall quality of movies is average, with a few exceptional or poor films.
