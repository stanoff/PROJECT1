**DATA GATHERING **

Since we are using python programming language for data science we need to import python packages that will be utilized in this study and then load the datasets relevant for our study.

```python
#python packages that will be utilized in this project
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#import 'bom.movie_gross.csv' from Box office Mojo data
df_movie_gross = pd.read_csv('bom.movie_gross.csv') #the index_col=0 is specified to avoid duplicated index
df_movie_gross
```

```python
#import 'title.basics.csv' from IMDB
df_title_basics = pd.read_csv('title.basics.csv',index_col = 0)
df_title_basics
```

```python
#import  'title.ratings.csv' from IMDB
df_ratings = pd.read_csv('title.ratings.csv',index_col = 0)
df_ratings
```

```python
#import data from the movie database
df_movies= pd.read_csv('tmdb.movies.csv',index_col = 0)
df_movies
```

**DATA CLEANING AND PREPARATION**

This involves checking data to check  if there exists any possible problems that makes data unsuitable for analysis.The problems may include missing data , duplicated data , outliers and inconsistencies in the data.Data cleaning and preparation done in our datasets include dropping rows with missing values, dropping unnecessary columns, getting rid of duplicates, creating new dataframes that suit our study including joining and merging datasets.Each datacleaning process is described in details in each cell based on the dataset.


```python
#Cleaning 'title.basics.csv'
#Drop unnecessary columns using drop()
df_title_basics.drop(['original_title','start_year','runtime_minutes'], axis = 1,inplace = True)
#Remove rows with missing genres
df_title_basics = df_title_basics[df_title_basics['genres'].notna()]
print(df_title_basics)
print(df_title_basics.head())
print(df_title_basics.info())
print(df_title_basics.describe())
```

```python
#Cleaning 'title.ratings.csv'
#Drop unnecessary columns
df_ratings.drop(['numvotes'],axis = 1,inplace = True)
print(df_ratings)
print(df_ratings.head())
print(df_ratings.info())
print(df_ratings.describe())
```

```python
#cleaning df_movie_gross
# Create a new dataframe with only 'domestic_gross' and 'title' columns
#remove duplicates
df_movie_gross_created= df_movie_gross[['title' , 'domestic_gross']]
print(df_movie_gross_created.drop_duplicates())
```

```python
#cleaning df_movies
# Create a new dataframe with only 'popularity' and 'title' columns
df_filtered = df_movies[['popularity', 'title']]
print(df_filtered)
```

```python
#join 'df_ratings' and 'df_title_basics' to get one dataframe 'df_created'
df_created = df_ratings.join(df_title_basics, on='tconst')
print(df_created)
print(df_created.info())
```

**EXPLORATORY DATA ANALYSIS**

This involves examining and understanding the structures, patterns and relationships within dataset.It also involves calculating basic descriptive statistics, creating  plots like bar graphs visualizations to focus on specific segments of data.Bar graph visualization suited this analysis and thus was employed in visualizing every aspect of data required.

```python
# split the genre column for analysis, so that each genre is independent
moviedf_small = df_created.assign(genres = df_created['genres'].str.split(',')).explode('genres')
print(moviedf_small)
```

```python
# Average rating per genre
average_rating_genre = moviedf_small.groupby('genres')['averagerating'].mean()
average_rating_genre.plot(kind='bar', figsize=(10, 6))
plt.title('Average Rating per Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.show()
```

```python
df_created.rename(columns = {'primary_title':'title'},inplace = True)
df_created.dropna(inplace = True)
print(df_created)
```

```python
# split the genre column for analysis, so that each genre is independent
moviedf_big = df_created.assign(genres = df_created['genres'].str.split(',')).explode('genres')
print(moviedf_big)
```

Below is a graph showing the domestic gross earnings per genre:
    

```python
#Domestic gross per genre
df_movie_gross_merged = moviedf_big.merge(df_movie_gross_created, on='title', how='inner')
domestic_gross_genre = df_movie_gross_merged.groupby('genres')['domestic_gross'].sum()
domestic_gross_genre.plot(kind='bar', figsize=(10, 6))
plt.title('Domestic Gross per Genre')
plt.xlabel('Genre')
plt.ylabel('Total Domestic Gross(in billion)')
plt.show()
```

```python
#join 'df_filtered' and 'df_created' to get one dataframe 'df_popularity'
df_popularity = df_filtered.merge(df_created, on=['title'],suffixes=('_left', '_right'))
print(df_popularity)
```

```python
# split the genre column for analysis, so that each genre is independent
moviedf_bigger= df_popularity.assign(genres = df_popularity['genres'].str.split(',')).explode('genres')
print(moviedf_bigger)
```

Below is a graph showing movie popularity per genre:
    

```python
#popularity per genre
popularity_genre = moviedf_bigger.groupby('genres')['popularity'].sum()
popularity_genre.plot(kind = 'bar',figsize = (10,6))
plt.title('Popularity per Genre')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.show()
```

-Action movies genre and adventure genre had relatively high domestic gross earnings.News,War and Western genres had the least domestic gross earnings.
