```python
#Cleaning 'title.basics.csv'
#Drop unnecessary columns using drop()
df_title_basics.drop(['original_title','start_year','runtime_minutes'], axis = 1,inplace = True)
#Remove rows with missing genres
df_title_basics = df_title_basics[df_title_basics['genres'].notna()]
print(df_title_basics)
print(df_title_basics.head())
print(df_title_basics.info())
print(df_title_basics.describe())
```

```python
#Cleaning 'title.ratings.csv'
#Drop unnecessary columns
df_ratings.drop(['numvotes'],axis = 1,inplace = True)
print(df_ratings)
print(df_ratings.head())
print(df_ratings.info())
print(df_ratings.describe())
```

```python
#cleaning df_movie_gross
# Create a new dataframe with only 'domestic_gross' and 'title' columns
#remove duplicates


df_movie_gross_created= df_movie_gross[['title' , 'domestic_gross']]
print(df_movie_gross_created.drop_duplicates())
```

```python
#cleaning df_movies
# Create a new dataframe with only 'popularity' and 'title' columns
df_filtered = df_movies[['popularity', 'title']]
print(df_filtered)
```

```python
#join 'df_ratings' and 'df_title_basics' to get one dataframe 'df_created'
df_created = df_ratings.join(df_title_basics, on='tconst')
print(df_created)
print(df_created.info())
```

```python
# split the genre column for analysis, so that each genre is independent
moviedf_small = df_created.assign(genres = df_created['genres'].str.split(',')).explode('genres')
print(moviedf_small)
```

```python
# Average rating per genre
average_rating_genre = moviedf_small.groupby('genres')['averagerating'].mean()
average_rating_genre.plot(kind='bar', figsize=(10, 6))
plt.title('Average Rating per Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.show()
```

```python
df_created.rename(columns = {'primary_title':'title'},inplace = True)
df_created.dropna(inplace = True)
print(df_created)
```

```python
# split the genre column for analysis, so that each genre is independent
moviedf_big = df_created.assign(genres = df_created['genres'].str.split(',')).explode('genres')
print(moviedf_big)
```

```python
#Domestic gross per genre
df_movie_gross_merged = moviedf_big.merge(df_movie_gross_created, on='title', how='inner')
domestic_gross_genre = df_movie_gross_merged.groupby('genres')['domestic_gross'].sum()
domestic_gross_genre.plot(kind='bar', figsize=(10, 6))
plt.title('Domestic Gross per Genre')
plt.xlabel('Genre')
plt.ylabel('Total Domestic Gross')
plt.show()
```

```python
#join 'df_filtered' and 'df_created' to get one dataframe 'df_popularity'
#df_popularity = df_filtered.join(df_created, on=['title'])
#print(df_popularity)
```

```python
#Create a new dataframe with the count of movies per genre
#genre_counts = df_imdb['genres'].value_counts().reset_index()
#genre_counts.columns = ['genres', 'count']
#print(genre_counts)

genre_counts = (pd.DataFrame(moviedf_small.groupby('genres', as_index = False).title.nunique())).sort_values(
    by = 'title', ascending = False)
print(genre_counts)
```

Below is a histogram showing the distribution of movie ratings


```python
#Histogram of movie ratings
plt.figure(figsize = (8,6))
sns.histplot(data = df_imdb, x = 'averagerating',bins = 20,kde =True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
